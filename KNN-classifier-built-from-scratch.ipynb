```python
import numpy as np
import statistics
```

```python
class KNN_classifier():
  def __init__(self,distance_metrics):
    self.get_distance_metrics=distance_metrics


  def get_distance_metrics(self,training_data_point,test_data_point):
    if self.distance_metrics=='euclidean':
      dist=np.sqrt(np.sum((training_data_point-test_data_point)**2))
      return dist

    elif self.distance_metrics=='manhattan':
      dist=np.sum(np.abs(training_data_point-test_data_point))
      return dist


  def nearest_neighbour (self,X_train,test_data,k):
    distance_list=[]
    for training_data in X_train:
      distance=self.get_distance_metrics(training_data,test_data)
      distance_list.append(training_data,distance)
      distance_list.sort(key=lambda x:x[1])

      neighbour_list=[]

      for j in range (k):
        neighbour_list.append(distance_list[j][0])

      return neighbour_list


  def predict(self,X_train,test_data,k):

    neighbors=self.nearest_neighbour(X_train,test_data,k)
    for data in neighbors:
      label=[]
      label.append(data[-1])
    predicted_class=statistics.mode(label)
    return predicted_class




```
